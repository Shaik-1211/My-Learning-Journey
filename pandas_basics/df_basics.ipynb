In the cell below, a Series with a daterange of 12 months is created and a dataframe is created with 3 columns namely Sales,Production,Price. The index of the dataframe is order_date indicating the sales of the month.

```python
import pandas as pd
import numpy as np
s = pd.Series(pd.date_range('20231101',periods = 12))
sales = pd.DataFrame(
    np.random.randint(1,10000,size = (12,4)),
    columns=['Sales','Production','Order_date','Price'])
sales['Order_date'] = s
sales.index = sales['Order_date']
sales = sales.drop('Order_date',axis=1)
sales
```

```python
# Creating a dataframe with different datatype columns.
df = pd.DataFrame({
    'S.No' : pd.Series(range(1,6)),
    'Date' : pd.Timestamp(20221103),
    'Temperature' : np.random.uniform(30.0,50.0,5),
    'WindSpeed':np.array(np.random.randint(52,254,5)),
    'Rained' : pd.Categorical(['True','False','False','True','True']),
    'City': 'Rayachoty'
})
df.index = df['S.No']
df = df.drop('S.No',axis=1)
df
```

```python
sorted = df.sort_index(ascending=False)#sort by index
temp_sort = df.sort_values(by = 'Temperature')#sort by values 
temp_sort
```

Accessing data by labels (by methods like .loc,.iloc,.st,conditional)

```python
value_by_loc = df.loc[0:3,['Temperature','WindSpeed','Rained']]
value_by_at=df.at[2,'WindSpeed']
value_by_iloc = df.iloc[0:5,0]
value_by_boolean = df[df['WindSpeed'] > 220]
print(value_by_boolean)
```
