Fillna and Dropna mehods are used in this file to handle missing data

```python
import pandas as pd 

df = pd.read_csv('student.csv')
df.fillna({
    'id' : 0,
    'class' : 10,
    'maths': 35,
    'science': 35,
    'social': 35,
    'status' : 'no status' 
})
```

```python
# forward fill method in fillna
new_df = df.fillna(method='ffill')
```

```python
#backward fill method 
new_df = df.bfill()
new_df
```

```python
# axis method in fillna
new_df = df.fillna( method='bfill',axis='columns')
new_df
```

```python
new_df = df.interpolate()
new_df
```

```python
new_df = df.dropna()
new_df
```

```python
df = pd.read_csv('student.csv')
df.set_index('Name',inplace=True)
new_df=df.dropna(thresh=6)
new_df
```
