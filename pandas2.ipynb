Reading and Writing data from a csv file and using some additional operations of `read_csv` method in pandas.

```python
import pandas as pd

df = pd.read_csv('annex.csv')
df
```

```python
df = pd.read_csv('annex.csv', skiprows = 1)
df
```

```python
df = pd.read_csv('annex.csv',skiprows = 1, nrows = 5)
df
```

```python
df = pd.read_csv('annex.csv',skiprows=1,nrows = 10)
df
```

```python
df = pd.read_csv('annex.csv',skiprows =1,nrows = 10,na_values=['not available','n.a'])
df
```

```python
df = pd.read_csv('annex.csv',skiprows=1,nrows=10)
df
```

```python
# Date can't be a negative integer so we have to convert it to NaN
df = pd.read_csv('annex.csv',skiprows=1,nrows = 10,na_values={
    'Date' : [-1,-2], # If these values(-1,-2) occur in date column convert it into NaN
    'Time' : [-26]
})
df
```

```python
#write dataframe to a csv file
df = pd.read_csv('annex.csv',skiprows=1,nrows=12,na_values={
    'Date' : [-1,-2], # If these values(-1,-2) occur in date column convert it into NaN
    'Time' : [-26]
})
df.to_csv('new_annex.csv',index = False)
```
