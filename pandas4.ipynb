Replace function is used to handle missing data

```python
import pandas as pd

df = pd.read_csv('pandas4.csv')
# original dataframe
df
```

```python
new_df = df.replace({
    'temperature' : '[A-Za-z]',
    'windspeed' : '[A-Za-z]'
},'',regex=True)
# Data frame after replacing unusual data from temperature and windspeed columns
new_df
```

```python
emp = pd.DataFrame({
    'dept' : ['CSE','ECE','Mechanical','Aeronautics','Civil'],
    'emp_names' : ['John','Ruhaan','Inara','Uzair','Anam']
})
emp
```

```python
emp = emp.replace(['CSE','ECE','Mechanical','Aeronautics','Civil'],['Cs','E','M','A','C'])
emp
```
