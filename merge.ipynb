```python
import pandas as pd 

weather_data = pd.read_csv('weather_data.csv')
date = pd.DataFrame(weather_data['day'])
city = pd.DataFrame(weather_data['city'])
left = pd.concat([date,city],axis=1)
left
```

```python
city = pd.DataFrame(weather_data['city'])
temp = pd.DataFrame(weather_data['temperature'])
right = pd.concat([city,temp],axis=1)
right
```

```python
new_df = pd.merge(left,right,on='city')#by default merge performs inner join
new_df
```

```python
df1 = pd.DataFrame({
    'city' : ['Newyork','paris','London','Delhi','Bangalore','Kanpur'],
    'temperature': [50,35,40,40,30,33]
})
df2 = pd.DataFrame({
    'city' : ['Newyork','Venice','London','Delhi','Bangalore','Hyderabad'],
    'windspeed':[45,50,60,32,88,50]
})
# Perform an left outer join using df1 and df2
merged = pd.merge(df1, df2,on='city',how='left')
merged
```

```python
# perform a right outer join using df1 and df2
right_outer = pd.merge(df1, df2,how='right')
right_outer
```

```python
#outer join
outer_join = pd.merge(df1, df2,how="outer")
outer_join
```
