```python
import pandas as pd
products = pd.DataFrame({
    'Name' : ['Chocolate','Biscuits','Chips','Rice','Oats'],
    'MarkedPrice': [100,50,50,250,100]
})
products 
```

```python
# Append another colums soldprice to products dataframe without axis keyword
price = pd.Series([75,50,50,200,75])
pd.concat([products,price])
```

```python
# Append another colums soldprice to products dataframe with axis keyword
price = pd.Series([75,50,50,200,75])
pd.concat([products,price],axis=1)
```
